```
import xlrd
import numpy as np
from pytube import YouTube
import os

from moviepy.editor import *
from moviepy.video.tools.segmenting import findObjects
```

```
book = xlrd.open_workbook('PRUEBA.xlsx')
sheet = book.sheet_by_name('Hoja1')
data = np.array([[sheet.cell_value(c, r) for c in range(sheet.nrows)] for r in range(sheet.ncols)])
# P

print(data)
```
